# Self made Decision Tree Classifier

```python
class Node:
  def __init__(self):
      Node.children={}
      Node.child=[]
      Node.col=None
class Tree:
    def __init__(self,df,pred, maxdepth=100):
        self.outcomes=df[pred].unique()
        self.maxdepth=maxdepth
        df=self.split(df)
        self.root=self.grow(df,pred)
    def gain(self,df,feature,pred,entropy_init):
        entropysum=0
        n=len(df.index)
        for j in df[feature].unique():
            X=df[df[feature]==j]
            e=self.entropy(X[pred])
            entropysum+=e*len(X)/n
        return entropy_init-entropysum
    def entropy(self,y):
        counts=y.value_counts()
        p=len(y)
        ans=0
        for i in self.outcomes:
            if i in counts:
                ans+=-(counts[i]/p)*math.log2(counts[i]/p)
        ans/=len(self.outcomes)
        if (ans<0.125):
            return 0
        return ans
    def true(self,y):
        e=self.entropy(y)
        if (e==1 or e==0):
            return True
        return False
    def findsplit(self,df,pred):
        if (self.true(df[pred])):
            for i in df[pred]:
                return i
        columns=df.columns[:-1]
        size=len(df)
        maxgain=0
        split_col=columns[0]
        e=self.entropy(df[pred])
        for i in columns:
            g=self.gain(df,i,pred,e)
            if (maxgain<g):
                split_col=i
                maxgain=g
        return split_col
    def grow(self,df,pred,depth=0):
        n=Node()
        col=self.findsplit(df,pred)
        if (col not in df.columns):
            return col
        dict={}
        n.col=col
        for i in df[col].unique():
            X=df[df[col]==i]
            if (depth==self.maxdepth):
                return i
            dict[i]=self.grow(X,pred,depth+1)
        n.children=dict
        return n
    def split(self,df):
        for i in df.columns:
            if (type(df[i].unique()[0])==str):
                continue
            part=(min(df[i])+max(df[i]))/3
            df[i]=df[i].apply(lambda x: "Low" if (x<part) else "Mid" if (x<part*2) else  "High")
        return df
    def predict(self,df):
        predictions=[]
        df=self.split(df)
        for index,row in df.iterrows():

            n=self.root
            while (n and type(n)!=str):
                cur=row[n.col]
                n=n.children[cur]
            predictions.append(n)
        return predictions
```

# Learning Overfitting on random dataset

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acc
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot
```

```python
X,y=make_classification(n_samples=10**4,n_features=20,n_informative=5,n_redundant=15, random_state=58)
print(X.shape,y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=58)
clf=DecisionTreeClassifier(criterion="entropy")
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print("Accuracy: ",acc(y_test,y_pred))
```

```python
train_accuracy=[]
test_accuracy=[]
X,y=make_classification(n_samples=10**4,n_features=20,n_informative=5,n_redundant=15, random_state=58)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=58)
for x in range(1,20):
    clf=DecisionTreeClassifier(max_depth=x,criterion="entropy")
    clf=clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    test_a=acc(y_test,y_pred)
    test_accuracy.append(test_a)
    y_pred=clf.predict(X_train)
    train_a=acc(y_train,y_pred)
    train_accuracy.append(train_a)
    print(x,train_a,test_a)
```

```python
plt.plot(train_accuracy,"-o")
plt.plot(test_accuracy,"-o")
plt.show()
```

```python
clf=DecisionTreeClassifier(max_depth=10,criterion="entropy")
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print("Accuracy: ",acc(y_test,y_pred))
```

# Applying the overfitting concept and making the best tree for Breast Cancer Dataset

```python
cancer=pd.read_csv("Breast_cancer_data.csv")
cancer.head()
```

```python
features=["mean_radius","mean_texture","mean_perimeter","mean_area","mean_smoothness"]
X=cancer[features]
y=cancer.diagnosis
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=58)
train_accuracy=[]
test_accuracy=[]
values=[i for i in range(1,10)]
for i in values:
    clf=DecisionTreeClassifier(max_depth=i)
    clf=clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    test_accuracy.append(acc(y_test,y_pred))
    y_pred=clf.predict(X_train)
    train_accuracy.append(acc(y_train,y_pred))
```

```python
plt.plot(values,train_accuracy,"-o",label="train")
plt.plot(values,test_accuracy,"-o",label="test")
plt.legend()
plt.show()
```

```python
clf=DecisionTreeClassifier(max_depth=6)
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print("Accuracy: ",acc(y_test,y_pred))
```
